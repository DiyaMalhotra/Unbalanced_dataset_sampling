```python
#cluster sampling
import random as rd
clusters=5
df1=df
state=None
length = len(df1)
element_max = length / clusters
  
    
cluster_list = []
cluster_id = 0
element_count = 0

for _ in df.iterrows():
      cluster_list.append(cluster_id)
      element_count += 1
      if element_count > (element_max - 1):
          element_count = 0
          cluster_id += 1
    
df1['cluster'] = cluster_list
  
rd.seed(state)
cluster_selected = rd.randint(0, clusters - 1)
  
dataframe_clustered = df1[df1['cluster'] == cluster_selected]
  
S4=dataframe_clustered.iloc[:,:-1]
S4
```

```python
#weighted sampling
def get_weighted_sample(df,n):
    def get_class_prob(x):
        weight_x = int(np.rint(n * len(x[x.Class != 0]) / len(df[df.Class != 0])))
        sampled_x = x.sample(weight_x).reset_index(drop=True)
        return (sampled_x)
        # we are grouping by the target class we use for the proportions

    weighted_sample = df.groupby('Class').apply(get_class_prob)
    print(weighted_sample["Class"].value_counts())
    return (weighted_sample)

S5= get_weighted_sample(df,100)
S5
```

```python
#defining model 1
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
def M1(sample):
    X=sample.iloc[:,:-1]
    Y=sample.iloc[:,-1]
    x_train, m, y_train, n= train_test_split(X,Y, test_size=0.33, random_state=42)
    clf = QuadraticDiscriminantAnalysis()
    clf.fit(x_train, y_train)
    clf.predict(X_test)
    return clf.score(X_test, y_test)
```

```python
#defining model 2
from sklearn.ensemble import ExtraTreesClassifier

def M2(sample):
    X=sample.iloc[:,:-1]
    Y=sample.iloc[:,-1]
    x_train, m, y_train, n= train_test_split(X,Y, test_size=0.33, random_state=42)
    clf = ExtraTreesClassifier(n_estimators=100, random_state=0)
    clf.fit(x_train, y_train)
    clf.predict(X_test)
    return clf.score(X_test, y_test)
```

```python
import lightgbm as lgb
from lightgbm import LGBMClassifier
def M3(sample):
    X=sample.iloc[:,:-1]
    Y=sample.iloc[:,-1]
    x_train, m, y_train, n= train_test_split(X,Y, test_size=0.33, random_state=42)
    clf = LGBMClassifier()
    clf.fit(x_train, y_train)
    clf.predict(X_test)
    return clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
def M4(sample):
    X=sample.iloc[:,:-1]
    Y=sample.iloc[:,-1]
    x_train, m, y_train, n= train_test_split(X,Y, test_size=0.33, random_state=42)
    clf = RandomForestClassifier(max_depth=2, random_state=0)
    clf.fit(x_train, y_train)
    clf.predict(X_test)
    return clf.score(X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
def M5(sample):
    X=sample.iloc[:,:-1]
    Y=sample.iloc[:,-1]
    x_train, m, y_train, n= train_test_split(X,Y, test_size=0.33, random_state=42)
    clf = DecisionTreeClassifier(random_state=0)
    clf.fit(x_train, y_train)
    clf.predict(X_test)
    return clf.score(X_test, y_test)
```
